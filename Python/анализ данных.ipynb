Вы работаете аналитиком в торговой сети и изучаете суммарную прибыль по каждой из категории товаров. Найдите товар, приносящий максимальную суммарную прибыль. 

ФОРМАТ ВВОДА

- Записи о продаже товаров: название товара и, через запятую, полученная с этой продажи прибыль (целое число). Каждая запись вводится с новой строки. Название товара может содержать любые символы, кроме запятой.

- Один товар могут продавать любое количество раз, но в каждой строке указывается информация только про одну продажу.

- Когда записи заканчиваются, на отдельной строке вводится слово "END". 


ФОРМАТ ВЫВОДА

- Название товара, который принес наибольшую суммарную прибыль и, через пробел, сумма этой прибыли.
- Гарантируется, что такой товар только один.
- Для примера:

Ввод
	
Шоколад,1000
Мармеладки,600
Шоколад,6000
Киндер-сюрприз,1000
Мармеладки,300
END

Шоколад,1000
END


Результат

Шоколад 7000

Шоколад 1000

```python
product = input()
data = {}
while product != 'END':
    product = product.split(',')
    if product[0] not in data:
        data[product[0]] = 0
    data[product[0]] += int(product[1])
    product = input()
    
data = dict(sorted(data.items(), key = lambda x: -x[1]))

print(*list(data.items())[0], sep=' ')
```

Вы планируете посетить концерт одной из своих любимых групп во время летнего отпуска. Найдите группу, у которой будет наибольшее количество концертов в 2023 году.


ФОРМАТ ВВОДА

- Записи о концертах: название группы и, через запятую с пробелом, дата концерта в формате ДД.ММ.ГГГГ. Каждая запись вводится с новой строки. Название группы может содержать любые символы, кроме запятой и точки.

-У каждой группы может быть любое количество концертов, но в каждой строке указывается информация только про один концерт.

-Когда записи заканчиваются, на отдельной строке вводится слово "END".

ФОРМАТ ВЫВОДА

- Название группы с наибольшим количеством концертов в 2023 году и, через пробел, количество концертов.
- Гарантируется, что такая группа только одна. 

Для примера:

Ввод	
Агата Кристи, 23.02.2024
Веселые программисты, 19.06.2023
Агата Кристи, 25.02.2023
Питоняшки, 27.08.2023
Питоняшки, 04.05.2023
Питоняшки, 27.02.2024
END

Питоняшки, 24.03.2024
Питоняшки, 21.05.2024
Глас Питона, 23.05.2023
Глас Питона, 20.09.2023
Би-2, 31.03.2023
Глас Питона, 24.12.2023
END


Результат

Питоняшки 2

Глас Питона 3

```python
info = input()
data = {}

while info != 'END':
    info = info.split(', ')
    if info[1][-4::] == '2023':
        if info[0] not in data:
            data[info[0]] = 0
        data[info[0]] += 1
    info = input()

data = dict(sorted(data.items(), key = lambda x : -x[1]))

print(*list(data.items())[0], sep=' ')
```
