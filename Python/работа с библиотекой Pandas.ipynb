```python
import numpy as np
data = [[(4,4),(5,2),(6,-2)], [(1,7),(8,5),(9,-5)], [(10,5),(11,3),(12,0)]]
df = pd.DataFrame(data, columns=['D','E','F'], index=['A', 'B', 'C']) 
lst1, lst2, maxim1, maxim2 = [], [], [], []

for lst in df.values:
    for tp in lst:
        lst1.append(tp[0])
        lst2.append(tp[1])
    maxim1.append((max(lst1), max(lst2)))
   
lst1.clear(), lst2.clear()

for lst in df.values.T:
    for tp in lst:
        lst1.append(tp[0])
        lst2.append(tp[1])
    maxim2.append((max(lst1), max(lst2)))

new_list = [maxim2]
new_row = pd.DataFrame(new_list, columns=['D', 'E', 'F'], index=['Best'])

df['Best'] = maxim1
df_new = pd.concat([df, new_row])
df_new 
```

```python
df = pd.DataFrame({
    'Color': ['Red', 'Green', 'Blue', 'Green']})

dummies = pd.get_dummies(df['Color'], dtype=int)

codes, uniques = pd.factorize(df['Color'])

df['Colors restored'] = codes
df['Colors restored'] = df['Colors restored'].map(lambda x: uniques[x])

'''Гибкость map:

map может принимать различные типы аргументов:
Функция: Как в вашем случае, можно передать функцию (или lambda функцию), которая будет применяться к каждому элементу.
Словарь или Series: Если передать словарь, map будет заменять значения согласно ключам словаря.
Callable: Можно передавать другие вызываемые объекты, которые будут применяться к значениям.

map vs apply:

map: Применяется только к объектам типа Series и работает поэлементно.
apply: Может применяться как к Series, так и к DataFrame, и может работать более гибко (например, по строкам или столбцам).'''
df
```

```python
codes, uniques = pd.factorize(df['Color'])
df['Colors restored'] = codes
df['Colors restored'] = df['Colors restored'].map(uniques.__getitem__)
df
```

```python
codes, uniques = pd.factorize(df['Color'])
df['Colors restored'] = codes
color_mapping = {code: color for code, color in enumerate(uniques)}
print(color_mapping)
df['Colors restored'] = df['Colors restored'].map(color_mapping)
df
```
